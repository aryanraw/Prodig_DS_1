# Total Population of India

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

loading dataset

```python
df = pd.read_csv("/content/world total (population).csv",encoding='utf-8')
df.head()
```

cleaning dataset

```python
df_final = df.drop(['Country Code','Indicator Name','Indicator Code'],axis=1)
```

ploting bar graph

```python
chosen_country = 'India'
country_data = df_final[df_final['Country Name'] == chosen_country]
years = country_data.columns[2:].astype('string')
population = country_data.iloc[:, 2:].values.flatten()
plt.figure(figsize=(9,3))
plt.bar(years, population, color='skyblue', edgecolor='black')
plt.xlabel('Year')
plt.ylabel('Population')
interval = 5
plt.gca().xaxis.set_major_locator(MultipleLocator(interval))
plt.title(f'Population of {chosen_country} Over the Years')
plt.show()
```

ploting histogram

```python
plt.hist(population, bins=20, color='skyblue', edgecolor='black')
plt.show()
```
